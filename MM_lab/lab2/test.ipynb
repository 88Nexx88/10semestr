```python
import matplotlib.pyplot as plt
import seaborn as sns

import main

sns.set()

import scipy.cluster.hierarchy as sch

print('\n\n\n\n')
dendrogram = sch.dendrogram(sch.linkage(main.df, method = 'single'))
print(sch.linkage(main.df, method = 'single'))
print(dendrogram)
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
dendrogram = sch.dendrogram(sch.linkage(main.df, method = 'complete'))

plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```
